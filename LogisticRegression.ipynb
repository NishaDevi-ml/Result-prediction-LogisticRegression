# LOGISTIC REGRESSION

```python
using dataframe
```

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df=pd.DataFrame([[2,0],[3,0],[4,0],[5,1],[6,1],[7,1],[8,1]],columns=["studyhours","result"])
df
```

```python
X=df[["studyhours"]]
Y=df["result"]
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR=LogisticRegression()
```

```python
LR.fit(X,Y)
```

```python
y=LR.predict([[4]])
y
```

```python
x=np.linspace(X.min(),X.max(),100)
y=LR.predict_proba(x)[:,1]
```

```python
plt.title('logistic Regression')
plt.scatter(X,Y)
plt.xlabel("study Hours")
plt.ylabel('Result')
plt.plot(X.values,LR.predict(X))
plt.plot(x,y)
```
